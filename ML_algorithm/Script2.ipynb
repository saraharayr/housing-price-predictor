```python
# Dependencies

import numpy as np
import pandas as pd
from pathlib import Path
```

```python
# Import the combined dataset

combined_df = pd.read_csv(Path('combined_df_updated.csv'))
combined_df = combined_df.drop(['Unnamed: 0'],axis='columns')
combined_df
```

```python
combined_df.info()
```

```python
# Convert the required catagorical columns into numerical data

property_type_dict = {'F':0,'O':1,'T':2,'S':3, 'D':5}
df1 = combined_df.replace({'property_type':property_type_dict})
Yes_No = {'Y':0,'N':1}
df2 = df1.replace({'new_build':Yes_No})
estate_type_dict = {'L':0, 'F':1}
df3 = df2.replace({'estate_type':estate_type_dict})
gender_type = {'Mixed':0, 'Girls':1, 'Boys':2}
df4 = df3.replace({'GENDER':gender_type})
df4
```

```python
# Drop any duplicates within the dataset

df5 = df4.drop_duplicates(subset=None, keep='first', inplace=False, ignore_index=False)
df5
```

```python
# Group the postcodes together

pc = df5["POSTCODE"].str.split(pat=' ', expand=True)
new_postcode = pc.drop([1], axis=1)
new_postcode
Pc2 = new_postcode[0].str.replace('B', '')

Pc2.value_counts()
```

```python
# Form a cleaner DataFrame with only the required columns
clean_df = pd.DataFrame({'Postcode': Pc2,
                         'OFSTED Rating': df5['OFSTEDRATING'],
                         'Price Paid': df5['price_paid'],
                         'Property Type': df5['property_type'],
                         'New Build': df5['new_build'],
                         'Estate Type': df5['estate_type'],
                         'Bus Route Number': df5['Bus Route Number']})
clean_df
```

```python
# Convert Postcode from an object to an integer
clean_df['Postcode'] = pd.to_numeric(clean_df['Postcode'],errors = 'coerce')
clean_df.info()
```

```python
# Split the data
data = clean_df.drop(['Price Paid'], axis=1)
target = clean_df['Price Paid'].values

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data, target, random_state=42)
```

# Unscaled Data Predictions

Our prediction would be that the Random Forest Classifier will work best for our dataset in comparison to Logistic Regression as it tends to have high accuracy prediction as well as being able to handle a large amount of features.

```python
# Train the Logistic Regression model on the unscaled data

from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(solver='lbfgs', max_iter=100000)
clf
```

```python
# Fitting the trained data and print the model score
clf.fit(X_train, y_train)

print(f"Logistic Regression Training Data Score: {clf.score(X_train, y_train)}")
print(f"Logistic Regression Testing Data Score: {clf.score(X_test, y_test)}")
```

```python
# Train a Random Forest Classifier model and print the model score

from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(random_state=42, n_estimators=500)
clf.fit(X_train, y_train)

print(f'RFC Training Score: {clf.score(X_train, y_train)}')
print(f'RFC Testing Score: {clf.score(X_test, y_test)}')
```

# Unscaled Data Results

The Random Forest Classifier achieved 54% for the training score whereas the testing score only achieved 38%, which was lower in comparison to the testing score of the Logistic Regression that achieved 45% thus disproving our hypothesis.

# Scaled Data Predictions

Random Forest Classifier is not as sensitive to scaling and so our prediction would be that its score should not be affected. However, the Logistic Regression score will improve significantly after scaling the data. In terms of which model should perform better, we assume RFC will have the higher accuracy scores.

```python
# Scale the data

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler().fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Train the Logistic Regression model on the scaled data and print the model score 

clf = LogisticRegression(solver='lbfgs', max_iter=10000)
clf.fit(X_train_scaled, y_train)

print(f"Logistic Regression Training Data Score: {clf.score(X_train_scaled, y_train)}")
print(f"Logistic Regression Testing Data Score: {clf.score(X_test_scaled, y_test)}")
```

```python
# Train a Random Forest Classifier model on the scaled data and print the model score

clf = RandomForestClassifier(random_state=42, n_estimators=500).fit(X_train_scaled, y_train)

print(f"RFC Training Data Score: {clf.score(X_train_scaled, y_train)}")
print(f'RFC Testing Data Score: {clf.score(X_test_scaled, y_test)}')
```

# Scaled Data Results

As predicted, after scaling the data, the Random Forest Classifier obtained the exact same scores as the unscaled data. In addition, the Logistic Regression model had a slight improvement on its training data score as expected with an increase of 0.4%.
Although the Random Forest Classifier testing scores did not prove to be as accurate as that of the Logistic Regression model, it did score higher on its training data scores. However, it was the Logistic Regression model that functioned better with our dataset overall as the difference between its testing and training scores were not as significant.
