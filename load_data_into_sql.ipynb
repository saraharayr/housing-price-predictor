# Loading all csv files into postgres

```python
#loading all required libraries
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
```

```python
#reading the data csv files
combined_schools_df=pd.read_csv('combined_schools_df.csv',index_col=[0])
busroutes_df=pd.read_csv('buses.csv', index_col=[0])
crimes_df=pd.read_csv('crimes_df_updated.csv',index_col=[0])
house_price_df=pd.read_csv('house_price_clean.csv',index_col=[0])
```

```python
#making a connect to the postgres 

rds_connection_string = "postgres:postgres@localhost:5432/project4"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
# displaying the table names
engine.table_names()
```

```python
#resetting index
combined_schools_df.reset_index(inplace=True)
```

```python
#renaming the index column
combined_schools_df.rename(columns={'index':'id'}, inplace=True)
```

```python
combined_schools_df.head()
```

```python
##load data into postgresql
combined_schools_df.to_sql(name='schools', con=engine, if_exists='replace', index=False)
```

```python
#resetting index
busroutes_df.reset_index(inplace=True)
```

```python
#renaming the index column
busroutes_df.rename(columns={'index':'id'}, inplace=True)
```

```python
##load data into postgresql
busroutes_df.to_sql(name='busroutes', con=engine, if_exists='replace', index=False)
```

```python
crimes_df.reset_index(inplace=True)
```

```python
#renaming the index column
crimes_df.rename(columns={'index':'id'}, inplace=True)
```

```python
##load data into postgresql
crimes_df.to_sql(name='crimes', con=engine, if_exists='replace', index=False)
```

```python
#resetting index
house_price_df.reset_index(inplace=True)
```
