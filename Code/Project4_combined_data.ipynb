# Merging the combined school data with bus routes, crimes and house pricing data

```python
schools_houseprice_df=pd.merge(combined_schools_df,house_price_df, how='inner', left_on='STREET', right_on='street')
```

```python
schools_houseprice_df = schools_houseprice_df[schools_houseprice_df['OFSTEDRATING'].notna()]
#schools_houseprice_df.dropna(subset=['OFSTEDRATING'])
```

```python
OFSTEDRATING_dict = {'Outstanding':1,'Good':2,'Requires improvement':3,'Special Measures':4,'Serious Weaknesses':5}
schools_houseprice_df_updated=schools_houseprice_df.replace({'OFSTEDRATING':OFSTEDRATING_dict})
```
